# Resource types -> plaintext formats (archive-to-JSON)

When converting an archive to JSON, each resource is stored either as **plaintext** (e.g. GFF->JSON, TLK->JSON) or as **base64** when no plaintext converter exists.

| Container | JSON schema | Notes |
|-----------|-------------|-------|
| ERF / MOD / SAV | `format: "erf"`, `erf_type`, `resources[]` | MOD/SAV use ERF layout |
| RIM | `format: "rim"`, `resources[]` | |
| BIF | `format: "bif"`, `bif_type`, `resources[]` with `resname_key_index` | |

| Resource type | Plaintext encoding in JSON |
|---------------|----------------------------|
| GFF, ARE, DLG, GIT, IFO, UTI, UTC, UTD, UTE, UTP, UTS, UTT, UTW, PTH, FAC, GUI, UTM, JRL | `gff_json` (nested dict) |
| TLK | `tlk_json` |
| 2DA | `2da_json` |
| LIP | `lip_json` |
| SSF | `ssf_xml` (string) |
| MDL, MDX, TGA, WAV, … | `base64` |

````python
# Registry used by archive_serializer (for reference)
from pykotor.tools.archive_serializer import _RESOURCE_PLAINTEXT_FORMAT

print("Resource types with plaintext embedding (restype -> plaintext format):")
for restype, text_format in sorted(
    _RESOURCE_PLAINTEXT_FORMAT.items(), key=lambda x: x[0].extension
):
    print(f"  {restype.extension} -> {text_format.extension}")
````

## CLI examples

```bash
# Archive -> JSON (plaintext resources embedded)
uv run pykotor capsule2json my.mod --output my.mod.json
uv run pykotor capsule2json data/player.bif --key-file chitin.key --output player.bif.json

# All resources as base64 (no plaintext)
uv run pykotor capsule2json my.erf --no-plaintext --output my.erf.json

# JSON -> capsule
uv run pykotor json2capsule my.mod.json --output my_restored.mod
```
